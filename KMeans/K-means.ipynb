# K-Means Clustering

### Import dependencies

```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np

# Create dummy data
from sklearn.datasets.samples_generator import make_blobs

# Kmeans model
from sklearn.cluster import KMeans
```

### Generate & plot random clusters

```python
X, _ = make_blobs(n_samples=500, centers=6, cluster_std=0.70, random_state=0)

# Plot and show scatter
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

### Create a K-Means model using `k=12`

```python
# Initialize the model
kmeans = KMeans(n_clusters=6)

# Fit the model to the data
kmeans.fit(X)

# Use the data to predict the clusters
# save the predictions as `predicted_clusters`
predicted_clusters = kmeans.predict(X)
predicted_clusters[:5]
```

### Plot the clusters

```python
plt.scatter(X[:, 0], X[:, 1], c=predicted_clusters, s=50, cmap='viridis')
plt.show()
```

### Capture the cluster centers

```python
centers = kmeans.cluster_centers_
labels = kmeans.labels_
centers
```

### Plot the data and the cluster centers

```python
plt.figure(figsize=(8, 8), dpi=80)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50)
plt.scatter(centers[:, 0], centers[:, 1], marker='*', c='red', s=150, alpha=0.95)

plt.show()
```

### View the migration of the cluster centers

```python
plt.figure(figsize=(20, 20), dpi=80)
counter=0

for n in range(1,16):
    kmeans = KMeans(n_clusters=6, init='random', precompute_distances=False, random_state=1234, n_init=1, max_iter=n)

    # Fit the model to the data
    kmeans.fit(X)

    # Use the data to predict the clusters
    # save the predictions as `predicted_clusters`
    predicted_clusters = kmeans.predict(X)
    
    # capture the cluster centers and their labels
    centers = kmeans.cluster_centers_
    labels = kmeans.labels_
    
    # Plot the clusters and their associated centers
    counter += 1
    ax = plt.subplot(5,3, counter)
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=50)
    plt.scatter(centers[:, 0], centers[:, 1], marker='*', c='red', s=150, alpha=0.95)
    plt.title(f"Number of iterations: {n}")
    
plt.show()
```
